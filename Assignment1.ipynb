# Factorial algorithms 

For the first part, we are going to compute the factorial algorithms by using three approaches: **iteractive** and **recursive** (tail and non-tail recursions). 
Next, we have to compute the running time as a function of n and compare the results for three approaches. 

## Iterative algorithm

An iterative algorithm will use repetition control structure. While looping statment to repeat the same steps. 
First, it will check if n is equal to 1 or 0, If yes, it will return 1. It will set variable to equal to n. While n is greater than 1, using a loop to keep multiplying variable by (n-1) until n is less than 1. And finally return the n-th factorial number as output.

### Pseudocode
    check if number is eqaul to 0 or equals to 1 
        
        if yes
        
            return 1 
        
        assign number to variable fact
        
        while number is greater than 1 
            
            variable multiply variable minus 1 
            
            number minus 1 
        
        return variable

### Implementation of Iterative algorithm

```python
def iterativeFactorial(n):
    """iteractive function for factorial function"""
    if n == 0 or n == 1:
        return 1
    fact = n
    while n > 1:
        fact *= (n-1)
        n-=1
    return fact
```

## Non-tail Recursive factorial algorithm

A non-tail recursive factorial algorithm will first check whether it reaches base case, if yes it returns 1. If not, it takes a natural number n, mulitiply the return value by (n - 1). And do the multiplication until it reach the base case (in this case 1). 

### Pseudocode
    
    A Function take number of positive integer
    
        if number eqauls to zero 
    
            return one  
        else 
            return number multiply Function itself with number minus 1 

### Implementation of Non-tail Recursive factorial algorithm

```python
def RecursiveFactorial(n):
    """Recursive version for factorial function"""
    if (n == 0):
        return 1
    else:
        return n * RecursiveFactorial(n-1)
```

## Tail recursive algorithm

Tail recursive algorithm is the tail recursive implementation of factorial. 
Two inputs will take which is n and an accumulator.
The accumulaotr is used for the multiplication and the function finishes at the tail.
To accumulate the value of factorial n and store it in the accumulator.
When n reaches 1, the accumulated value will be returned. The output of this algorithm is the n-th factorial number as output.

Tail-call optimization is a method which allows infinite recursion of tail-recursive functions to occur without stack overflow.

### Pseudocode
    
    A Function take number of positive integer and accumulator initialize to one
    
           if number is equal to one or zero
           
               return the accumulator 
               
           else
           
               return the function take the number minus 1 and accumulator multiply n 

### Implementation of Tail Recursive factorial algorithm

```python
def TailRecursiveFactorial(n,accumulator=1):
    """Tail Recursive version for factorial function"""
    if n ==1 or n ==0:
        return accumulator
    else:
        return TailRecursiveFactorial(n-1, n*accumulator)
```

### Running time for three approaches

```python
import matplotlib.pyplot as plt
import time 

def RunningTime(func):
    """Running time for function execution"""
    num = [n for n in range (1,500)]
    times= list()
    for i in num:
        #record start time 
        start = time.time()
        func(i)
        #end the record time
        stop = time.time()
        #append the result into a list
        times.append((stop - start)*1e6)
    return times

```

Supposedly, the running time of non-tail recursive should have the longest running time complexity, followed by tail recursive and iterative. 

This is becuase for non-tail recursive, each time the function calls itself a new frame is pushed on top of the previous frames until it reaches the base of the recursion (fact(0)). Then the base case returns 1 and its frame is popped off the call stack. Then fact(1) returns 1Ã—1 and its frame is popped. The other remaining frames of fact are popped one by one until fact(n) return the final value. 

For tail recursive, even though there is still function calls itself a new frame until it reach a base case, it does not need to do the multiplication back and simply return the accumulator after reaching the base case.

For iterative function, all the computation are done within a loop and it only returns one time. Since it doesn't call require functions being pushed in a call stack frames and return after it reached the base case. 

As shown in graph above, non-tail recursive factorial takes a longer time than tail-recursive approach until the size of input reaches 400. Tail Recrusive Approach took the longest running time when the number of size reaches 400.

# Time complexity of each approach

Time complexity is the number of operations an algorithm performs to complete its task (considering that each operation takes the same amount of time). 

The time complexity of each appraoch are the same. From the algorithm of factorials, time complexity are **O(n)** as the time computation proportional to the size of input. 



## Space complexity of each approach

Space complexity measures the total amount of memory that an algorithm or operation needs to run according to its input size.

For the **iterative algorithm**, there is no adding of new stack when function is called, so the whole algorithm take O(1) space for variable created. Therefore, the space complexity of the iterative algorithm is **O(1)**.

For the **tail recursive algorithm**, as no computation is performed on the returned value, there is no stack frame for function call. Therefore, the space complexity of the tail recursive algorithm is also **O(1)**.

For every call to the **non-tail recursive**, the state is saved onto the call stack, till the value is computed and returned to the called function. We don't assign an explicit stack, but an implicit call stack is maintained 
For factorial of N, a stack of size N will be impliciitly allocated for storing the state of the function calls. Therefore, the space complexity of the non-tail recursive algorithm is **O(n)**.

## Binary search algorithms running time with different approaches

This algorithm is an binary search algorithm using iterative approach. It will take a **sorted list** and an search element to be found as inputs. 

Iterative binary search algorithm takes the input of a sorted list and and an searched element. 
Inside the while loop, "mid" is obtained by calculating (low+high)/2.
If number at position mid equal target element then the control returns target and stop the loop 

Else, if target is less than the number at position mid, then we pass the number at position mid minus 1 to the upper level

Else if target is greater than the number at position mid, then we pass the number at position mid plus 1 to lower level.  

The while loop continues to iterate in this way till either the element is returned (indicating key has been found in the Array) or low becomes greater than high,in which case -1 is returned indicating that key could not be found. 

### Implementation 

```python
def iterativeBinary(lists,target):
    done = False
    upperLevel = len(lists) - 1 
    lowerLevel = 0
    while done is False:
        mid = (upperLevel + lowerLevel) // 2
        if target == lists[mid]: 
            return target
            done == True
        elif target < lists[mid]:
            upperLevel = mid - 1
        elif target > lists[mid]:
            lowerLevel = mid + 1
        if lowerLevel > upperLevel:
            break
    return -1

    
    
```

## Recursive binary search algorithm

This algorithm is an binary search algorithm using recursive approach. It will take a sorted list and an search element to be found as inputs.

Recursive binary search algorithm takes the input of a sorted list and and an searched element. Inside the If statement, "mid" is obtained by calculating (low+high)/2. If number at position mid equal target element then the control returns target and stop the loop

Else, if target is less than the number at position mid, then we pass the number at position mid minus 1 to the upper level. And return a function again with new upper level pass to it.

Else if target is greater than the number at position mid, then we pass the number at position mid plus 1 to lower level. And return a function again with new lower level pass to it.

The function continues to call itself in this way till either the element is returned (indicating key has been found in the Array) or low becomes greater than high,in which case -1 is returned indicating that key could not be found.

### implementation 

```python
def recursiveBinary(lists,target,upperLimit=None,lowerLimit=0):
    if upperLimit == None:
        upperLimit = len(lists)-1
    done = False
    mid = (upperLimit + lowerLimit) // 2
    #base case 
    if (lowerLimit > upperLimit):
        return -1
    if target == lists[mid]:
        return target
    #
    elif target < lists[mid]:
        upperLimit = mid - 1
        return recursiveBinary(lists,target,upperLimit,lowerLimit)
    elif target > lists[mid]:
        lowerLimit = mid + 1
        return recursiveBinary(lists,target,upperLimit,lowerLimit)
    
```

# 4. Curve fitting

In this part, we are going to use curve-fitting algorithms to find the best fit equation of the running times obtained from the previous parts. With the curve-fitting algorithms, we can see clearly that theoritically and ideally what the line-plotted should be like.

```python
from scipy.optimize import curve_fit
import numpy as np

def fit(x,a,b):
    return a*x+b

def plt_graph(name,result):
    xdata=np.array([n for n in range (1,500)])
    ydata=np.array(result)
    param1, covariance1 = curve_fit(fit, xdata, ydata)
    plt.figure()
    plt.plot(xdata,ydata,'b',label=name)
    #putting the * so that the plot can expand out the terms into the a, b that function is expecting.
    plt.plot(xdata,fit(xdata,*param1),'r-',label='fit: a=%5.3f, b=%5.3f' % tuple(param1))
    plt.title('Curve fitting: {0}'.format(name))
    plt.legend()
    plt.xlabel('n')
    plt.ylabel('Running time, microseconds')
#     plt.savefig('{0}.pdf'.format(name)) 
    plt.show()
```
